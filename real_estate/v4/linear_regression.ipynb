```python
import pandas as pd
import numpy as np
import statsmodels.api as sm

from sklearn.linear_model import LinearRegression

import matplotlib.pyplot as plt
% matplotlib inline
```

```python
test_df = pd.read_csv("~/real_estate/data/v4/final/yelp/yelp_test_df.csv")
train_df = pd.read_csv("~/real_estate/data/v4/final/yelp/yelp_train_df.csv")
train_df_adasyn = pd.read_csv("~/real_estate/data/v4/final/yelp/yelp_train_df_adasyn.csv")
train_df_smote_1 = pd.read_csv("~/real_estate/data/v4/final/yelp/yelp_train_df_smote_ratio.csv")
train_df_smote_2 = pd.read_csv("~/real_estate/data/v4/final/yelp/yelp_train_df_smote_ratio_2.csv")
train_df_smote_3 = pd.read_csv("~/real_estate/data/v4/final/yelp/yelp_train_df_smote_ratio_3.csv")

features = train_df.columns[:-2]

X_test = test_df[features].values
X_train = train_df[features].values
X_train_adasyn = train_df_adasyn[features].values
X_train_smote_1 = train_df_smote_1[features].values
X_train_smote_2 = train_df_smote_2[features].values
X_train_smote_3 = train_df_smote_3[features].values

y_test = test_df["target"].values
y_train = train_df["target"].values
y_train_adasyn = train_df_adasyn["target"].values
y_train_smote_1 = train_df_smote_1["target"].values
y_train_smote_2 = train_df_smote_2["target"].values
y_train_smote_3 = train_df_smote_3["target"].values
```

### Train Linear Regressions

```python
model_1 = LinearRegression()
model_1.fit(X_train, y_train)
y_test_hat_1 = model_1.predict(X_test)
y_train_hat_1 = model_1.predict(X_train)
print("R^2 train: ", np.square(np.corrcoef(y_train, y_train_hat_1)[0,1]))
print("R^2 test: ", np.square(np.corrcoef(y_test, y_test_hat_1)[0,1]))
```

### Plot Results

```python
plt.scatter(y_train, y_train_hat_1)
plt.title("Labels vs Predictions (In-Sample)")
plt.ylabel("Predictions")
plt.xlabel("Labels")
plt.axvline(x=0.25, color="red", linestyle='--')
plt.axhline(y=0.25, color="r", linestyle='--')
plt.show()

plt.scatter(y_test, y_test_hat_1)
plt.title("Labels vs Predictions (Out-of-Sample)")
plt.ylabel("Predictions")
plt.xlabel("Labels")
plt.axvline(x=0.25, color="red", linestyle='--')
plt.axhline(y=0.25, color="r", linestyle='--')
plt.show()
```

### Get Coefficients

```python
model = sm.OLS(y_train, X_train)
model = model.fit()

coefficients = model.params
standard_errors = model.HC0_se

regression_table = pd.DataFrame({"coefficients":coefficients, 
                                 "standard_errors":standard_errors}, index=features)
```
